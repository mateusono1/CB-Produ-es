// Função para remover colunas completamente vazias
function removeColunasVazias(data) {
  var colunasNaoVazias = [];
  var numLinhas = data.length;

  // Verifica cada coluna
  for (var col = 0; col < data[0].length; col++) {
    var coluna = data.map(function(row) { return row[col]; });
    if (coluna.some(function(cell) { return cell !== ""; })) {
      colunasNaoVazias.push(coluna); // Mantém a coluna se não for completamente vazia
    }
  }

  // Constrói a nova matriz com colunas não vazias
  var novaMatriz = [];
  for (var linha = 0; linha < numLinhas; linha++) {
    novaMatriz[linha] = [];
    for (var i = 0; i < colunasNaoVazias.length; i++) {
      novaMatriz[linha][i] = colunasNaoVazias[i][linha]; // Adiciona as colunas não vazias
    }
  }

  return novaMatriz;
}

// Função para remover linhas onde a primeira coluna é vazia
function removeLinhasPrimeiraColunaVazia(data) {
  return data.filter(function(row) {
    return row[0] !== ""; // Retorna true se a primeira coluna não estiver vazia
  });
}

// Função para remover colunas completamente vazias
function removeColunasVazias(data) {
  var colunasNaoVazias = [];
  var numLinhas = data.length;

  // Verifica cada coluna
  for (var col = 0; col < data[0].length; col++) {
    var coluna = data.map(function(row) { return row[col]; });
    if (coluna.some(function(cell) { return cell !== ""; })) {
      colunasNaoVazias.push(col);
    }
  }

  // Cria nova matriz com colunas não vazias
  return data.map(function(row) {
    return colunasNaoVazias.map(function(col) {
      return row[col];
    });
  });
}

  });
}

// Função para remover linhas completamente vazias
function removeLinhasVazias(dados) {
  return dados.filter(function(linha) {
    return linha.some(function(celula) {
      return celula !== "";
    });
  });
}

// Função para remover colunas completamente vazias
function removeColunasVazias(dados) {
  if (dados.length === 0) return dados; // Retorna se não há dados

  var colunasVazias = dados[0].map((_, colIndex) => dados.every(row => row[colIndex] === ""));
  return dados.map(row => row.filter((_, colIndex) => !colunasVazias[colIndex]));
}

// Função para remover linhas onde a primeira coluna está vazia
function removeLinhasPrimeiraColunaVazia(dados) {
  return dados.filter(function(linha) {
    return linha[0] !== ""; // Remove linhas onde a primeira coluna está vazia
  });
}

// Função para renomear as colunas relevantes
function renomearColunas(cabecalhos) {
  for (var i = 0; i < cabecalhos.length; i++) {
    if (cabecalhos[i] === 'data de nascimento ') {
      cabecalhos[i] = 'data de nascimento';
    }
  }
}
