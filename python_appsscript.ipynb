// Função para remover colunas completamente vazias
function removeColunasVazias(data) {
  var colunasNaoVazias = [];
  var numLinhas = data.length;

  // Verifica cada coluna
  for (var col = 0; col < data[0].length; col++) {
    var coluna = data.map(function(row) { return row[col]; });
    if (coluna.some(function(cell) { return cell !== ""; })) {
      colunasNaoVazias.push(coluna); // Mantém a coluna se não for completamente vazia
    }
  }

  // Constrói a nova matriz com colunas não vazias
  var novaMatriz = [];
  for (var linha = 0; linha < numLinhas; linha++) {
    novaMatriz[linha] = [];
    for (var i = 0; i < colunasNaoVazias.length; i++) {
      novaMatriz[linha][i] = colunasNaoVazias[i][linha]; // Adiciona as colunas não vazias
    }
  }

  return novaMatriz;
}

// Função para remover linhas onde a primeira coluna é vazia
function removeLinhasPrimeiraColunaVazia(data) {
  return data.filter(function(row) {
    return row[0] !== ""; // Retorna true se a primeira coluna não estiver vazia
  });
}

function removeColunasVazias(data) {
  var colunasNaoVazias = [];
  var numLinhas = data.length;

  // Verifica cada coluna
  for (var col = 0; col < data[0].length; col++) {
    var coluna = data.map(function(row) { return row[col]; });
    if (coluna.some(function(cell) { return cell !== ""; })) {
      colunasNaoVazias.push(coluna); // Mantém a coluna se não estiver completamente vazia
    }
  }

    var novaLinha = colunasNaoVazias.map(function(col) { return col[row]; });
    novoData.push(novaLinha);
  }
  
  return novoData;
}
